```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("data.csv")
data
```

```python
data=data.dropna()
data
```

```python
data["Actual"]=data["Actual"].replace(["FALSE", "TRUE", '0', '1'],[0, 1, 0, 1])
data["Predicted"]=data["Predicted"].replace(["FALSE", "TRUE", '0', '1'],[0, 1, 0, 1])
data
```

```python
#Get the confusion matrix
actual=[i for i in data["Actual"]]
predicted=[i for i in data["Predicted"]]
cf_matrix = confusion_matrix(actual, predicted)
print(cf_matrix)
```

```python
sns.heatmap(cf_matrix/np.sum(cf_matrix), fmt='.2%', annot=True)
plt.title("Confusion matrix of result obtained")
plt.ylabel("Actual")
plt.xlabel("Predicted")
```

```python
precision = precision_score(actual, predicted)
recall = recall_score(actual, predicted)
accuracy=accuracy_score(actual, predicted)
f1=(2*precision*recall)/(precision+recall)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
